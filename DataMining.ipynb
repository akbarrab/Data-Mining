```python
#Transformasi setiap elemen dalam setiap baris Data Frame df menjadi string dan menyimpannya sebagai list of lists
data = [[str(element) for element in row] for row in df.values]
data #pemanggilan
```

```python
#mengimpor TransactionEncoder dari library mlxtend
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder() #Membuat instance TransactionEncoder

data_encoded=te.fit(data).transform(data)# metode fit(data) mempelajari semua elemen unik dalam data. Lalu, transform(data) mengubah data transaksi ke dalam bentuk array NumPy

df = pd.DataFrame(data_encoded,columns=te.columns_)#membuat DataFrame df dari data_encoded
df = df.astype(int)#Mengonversi setiap nilai dalam df ke tipe int.
df#Menampilkan DataFrame
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
frequent_itemsets = apriori(df, min_support=0.0005, use_colnames=True)# Menjalankan apriori dengan min_support lebih rendah

rules = association_rules (frequent_itemsets, metric="lift", min_threshold=1.0)# Menjalankan association rules jika ada frequent itemsets yang ditemukan
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Ubah df ke tipe bool untuk meningkatkan performa dan menghilangkan peringatan
df = df.astype(bool)

# Jalankan apriori untuk menemukan frequent itemsets
frequent_itemsets = apriori(df, min_support=0.0005, use_colnames=True)

# Memastikan ada frequent itemsets sebelum melanjutkan
if not frequent_itemsets.empty:
    # Jalankan association rules
    rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

    # Tampilkan hasilnya
    print(rules[['antecedents', 'consequents', 'antecedent support', 'consequent support',
                 'support', 'confidence', 'lift', 'leverage', 'conviction', 'zhangs_metric']])
else:
    print("Tidak ada frequent itemsets yang ditemukan. Coba turunkan nilai min_support.")
```
